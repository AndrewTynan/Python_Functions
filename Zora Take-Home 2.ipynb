## Initial Feature Creation

```python
wallet_lifetime_days_df = (df 
                           .groupby('wallet_address') 
                           .agg(min_date = ('date','min'),
                                max_date = ('date','max')) 
                           .reset_index()
                           .pipe(lambda df_: pd.merge(df_, 
                                                      df.loc[:,['wallet_address', 'churned_at_date']].drop_duplicates(), 
                                                      how='left', 
                                                      on='wallet_address'))
                           .assign(lifetime_days          = lambda x: (x['max_date'] - x['min_date']).dt.days,
                                   is_churn_date_max_date = lambda x: np.where(x.churned_at_date == x.max_date, 'yes', 'no'),
                                   is_churned = lambda x: np.where(x.churned_at_date == x.churned_at_date, 'Yes', 'No'))) 

# print(wallet_lifetime_days_df.dtypes) 

# check that churn date is last active
(wallet_lifetime_days_df
.groupby('is_churn_date_max_date')
['wallet_address']
.count()
.reset_index()) 
```

```python
# get distinct wallet dates 
# and append wallet_lifetime_days_df
wallet_engagement_df = (collection_activity_df[['wallet_address', 'date']]
                        .drop_duplicates()
                        .dropna()
                        .sort_values(by = ['wallet_address', 'date']) 
                        .pipe(lambda df_: pd.merge(df_, 
                                                   wallet_lifetime_days_df, 
                                                   how = 'left', 
                                                   on = 'wallet_address'))
                        .reset_index()
                        .assign(prior_date               = lambda x: x.groupby('wallet_address')['date'].shift(1),
                                days_since_active        = lambda x: (x.date - x.prior_date).dt.days,
                                days_since_active_7d_avg = lambda x: x.groupby('wallet_address')['days_since_active']  
                                                                      .rolling(window=7, min_periods=1)
                                                                      .mean()
                                                                      .reset_index()
                                                                      .loc[:,'days_since_active'], # need to get just the last column 
                                active_count_7d          = (lambda x: x.groupby('wallet_address')['date']
                                                                        .rolling(window=7, min_periods=1)
                                                                        .count().astype(int)
                                                                        .reset_index()
                                                                        .loc[:,'date']),                                
                        ))

wallet_engagement_df.shape
```

```python
date_features_df = (df 
                      .groupby('wallet_address')
                      .agg(date_count    = ('date', 'count'),  # nulls will impact this .. 
                           unique_dates  = ('date', 'nunique'),
                           min_date      = ('date','min'),
                           max_date      = ('date','max'),
                           creator_count = ('creator','nunique'))
                      .reset_index())  

# NOTE collectors_df is one row per wallet_address
features_df = pd.merge(collectors_df, date_features_df, how='left', on='wallet_address')

wallet_address_df = (df
                    .loc[:, ['wallet_address']]
                    .drop_duplicates()) 
```

#### Chain Events
###### Get the counts of events by chain per wallet

```python
chains = list(df.dropna().chain_name.unique())
results = {} 

for chain in chains:
    agg = (df
           .query("chain_name == @chain")
           .groupby('wallet_address')
           .agg(chain_count = ('chain_name', 'count'))
           .reset_index())
    
    results[chain] = agg 

if 'wallet_chain_use_df' in globals(): del(wallet_chain_use_df) 
    
for chain in results:
    new_name = chain.lower() + '_count' 
    chain_results_df = results[chain].rename(columns = {'chain_count': new_name}) 

    if 'wallet_chain_use_df' in globals():
        wallet_chain_use_df = pd.merge(wallet_chain_use_df, chain_results_df, how='left', on='wallet_address') 
    else:
        wallet_chain_use_df = pd.merge(wallet_address_df, chain_results_df, how='left', on='wallet_address') 


features_df = pd.merge(features_df, wallet_chain_use_df, how='left', on='wallet_address')

features_df.head(1)
```

#### File Type Events
###### Get the counts of events by file type per wallet

```python
file_types = list(df.dropna().file_type.unique())
results = {} 

for file_type in file_types:
    agg = (df
         .query("file_type == @file_type")
         .groupby('wallet_address')
         .agg(file_type_count = ('file_type', 'count'))
         .reset_index())
    
    results[file_type] = agg 

if 'file_type_use_df' in globals(): del(file_type_use_df) 
    
for file_type in results:
    new_name = file_type.lower() + '_count' 
    file_type_results_df = results[file_type].rename(columns = {'file_type_count': new_name}) 

    if 'file_type_use_df' in globals():
        file_type_use_df = pd.merge(file_type_use_df, file_type_results_df, how='left', on='wallet_address') 
    else:
        file_type_use_df = pd.merge(wallet_address_df, file_type_results_df, how='left', on='wallet_address') 


features_df = pd.merge(features_df, file_type_use_df, how='left', on='wallet_address')

features_df.head(1)
```
