```python
import pandas as pd
pd.__version__
```

```python
data = {'Name': ['Giggs', 'Tom', 'Dick', 'Harry', 'Jack', 'Jill', 'Scholes', 'Martial', 'Rashford', 'Pogba'],
                'Age': [23,21,24,21,20,10,23,45,22,35],
                'Rating': [4.23, 3.21, 2.10, 1.91, 4.32, 6.32, 4.19, 2.09, 1.09, 3.33]}

df = pd.DataFrame(data) 

df['taco'] = df.apply(lambda x: x['Age']+3, axis='columns')
df
```

```python
def lol(df, x):
    
    output = []
    for i in df[x]: 
        if i > 3:
            output.append("High")
        else:
            output.append("Low")
            
    df[x + "_lolled"] = output
    
    return df


# use function 
df.pipe(lol, x = "Rating")
```

```python
def lol(df):
    if(df.Rating > 3):
        return "High"
    else:
        return "Low"

    
# use function 
df['hhm'] = df[['Rating']].apply(lol, axis='columns')

df
```

```python
def lol(df, col):
    if(df[col] > 3):
        return "High"
    else:
        return "Low"
    

# use function 
df['hhm'] = df.apply(lol, col = 'Rating', axis='columns') 

df
```

```python
def multi_conditions(x):
    if x<5:
        return "No need"
    elif x>=5 and x<=7:
        return "Hold decision"
    else:
        return 'Need'


# use function 
df['Maths_Spl Class'] = df['Rating'].apply(multi_conditions)

df
```

```python
# using a basic function
def df_cumsum_basic(df, var):
    var_name = var + "_CumSum"
    df[var_name] = df[var].cumsum()
    return df 

df_cumsum_basic(df, "Rating")    
```

```python
df.pipe(df_cumsum_basic, var="Rating")
```
