## define functions

```python
def get_region(state):
    """Returns the region of a given US state."""

    regions = {
        "Northeast": ["Maine", "New Hampshire", "Vermont", "Massachusetts", "Rhode Island", "Connecticut", "New York", "New Jersey", "Pennsylvania"],
        "Midwest": ["Ohio", "Indiana", "Illinois", "Michigan", "Wisconsin", "Minnesota", "Iowa", "Missouri", "North Dakota", "South Dakota", "Nebraska", "Kansas"],
        "South": ["Delaware", "Maryland", "Virginia", "West Virginia", "North Carolina", "South Carolina", "Georgia", "Florida", "Kentucky", "Tennessee", "Alabama", "Mississippi", "Arkansas", "Louisiana", "Oklahoma", "Texas"],
        "West": ["Montana", "Idaho", "Wyoming", "Colorado", "New Mexico", "Arizona", "Utah", "Nevada", "California", "Oregon", "Washington", "Alaska", "Hawaii"]
    }

    for region, states in regions.items():
        if state in states:
            return region

    return "Unknown"

# print(get_region("California")) 


# using a basic function
def df_cumsum_basic(df, var):
    var_name = var + "_CumSum"
    df[var_name] = df[var].cumsum()
    return df  


def lol(df, col):
    if(df[col] > 3):
        return "High"
    else:
        return "Low"
```

```python
def _conditions1(row):    
    if row["gpa"] > 3.8:
        if row["gpa"] >= 4.0: 
            rank = "A+"
        else:
            rank = "A"
    elif row["gpa"] >  3.2:     rank = "B"
    elif row["gpa"] <= 2.8:     rank = "C"
    else: rank = "F"
        
    return rank
```

```python
# chaining .assign functions to work with the new column 
(numbers_df
 .assign(new  = lambda x: x['Col1'] + x['Col2'],
         
         new2 = numbers_df['Col1'].apply(lambda x: 'Honors' if x > 3.5 else 'Normal'),  
         
         new22 = lambda x: x['Col1'].apply(lambda y: "High" if y >= 4 else ("Medium" if y >= 3 else "Low")),  
         
         # multi conditions, like case_when 
         huh = lambda x: np.select(condlist=(x.Col3 == 100,
                                            (x.Col3 > 100) & (x.Col3 < 500),
                                             x.Col3 >= 500),
                                     choicelist=('Low', 'Mid', 'High'),
                                     default='Other'),  
         
         huh2 = numbers_df.apply(_conditions1, axis=1),
         
         log_Col3 = lambda x: np.log(x.Col3),
         
         new3 = lambda x: x.apply(lambda y: y['Col1'] + y['Col2'], axis=1),
         Col5 = lambda x: x.new * 2,
         
         prdd = numbers_df.Col1 * numbers_df.Col2, 
         sqqt = np.sqrt(numbers_df.Col1), 
        
         now  = 'what',  
         test = np.where(numbers_df['new'] > 1000, "Yes", "No"),    
         text = np.where(lambda x: x['new'] > 1000, "Yes", "No"), 

        crazy = [1500 if x > 3 else 800 for x in numbers_df['new']],

        Category  = lambda x: 'Low' if x['newer'].isin(['Low']) is True else 'Literature',
                  
        Category2 = lambda x: x.newer.apply(lambda y: 'XX' if y in ['Low'] else 'Literature')          

        )
.loc[:, 'gpa':]
)
```

```python
def group_by_summarise(df, group_by, metric, stat): 

    col_name = "{metric}_{stat}".format(metric = metric, stat = stat)
    print(col_name)

    return (df
           .groupby(group_by)
           .agg(stat  = (metric, stat))
           .reset_index() 
           .rename(columns = {'stat': col_name}))


(group_by_summarise(df       = new_fake_df, 
                    group_by = ['state', 'region'], 
                    metric   = 'miles_driven',
                    stat     =  'mean')
.head(3))
```

```python
# more window functions 

# expanding(): Accumulating window over the values
# rolling():   Rolling window

newer_fake_df = (new_fake_df
                .assign(year = lambda x: x['date_time'].dt.strftime('%Y'))
                .groupby(['region', 'year']) 
                .agg(name_count = ('name', 'count')) 
                .reset_index())   
 
# note this is grouped by ['region', 'year']    
(newer_fake_df 
.assign(cumsum             = lambda x: x['name_count'].expanding().sum().astype(int), # same as cumsum() below 
        cumsumfunc         = lambda x: x['name_count'].cumsum(), 
        cum_max            = lambda x: x['name_count'].cummax(), 
        five_year_cumsum   = lambda x: x['name_count'].rolling(window=5, min_periods=1).sum().astype(int), # custom window
        five_year_centered = lambda x: x['name_count'].rolling(window=5, center=True).mean(), # custom window 
#         preceding_max      = lambda x: x['name_count'].rolling(window='unbounded preceding').sum(),
        
        row_number         = (newer_fake_df
                              .groupby('region')
                              .transform('cumcount') + 1), # count from 1 not 0  
     
        row_number_2       = (newer_fake_df
                              .groupby('region')['year']
                              .rank(method='first').astype(int)), # rank to make row_number 
        
        row_number_3       = (newer_fake_df
                              .groupby('region')['year']
                              .rank(pct=True, method='dense').astype(int)), # dense 
        
        cummin             = (newer_fake_df
                              .groupby('region')['name_count']
                              .transform('cummin')),   
        
#         cummin2             = (newer_fake_df
# #                               .groupby('region')
#                               .transform('cummin')),           
       )
.astype({'year': 'int'}) 
.query("year in [1970,2024]") 
)
```
